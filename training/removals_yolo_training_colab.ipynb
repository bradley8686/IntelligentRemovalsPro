## 2. Upload Dataset

Upload a YOLO-format dataset zip exported from your labelling tool. It should contain a `data.yaml` file.

```python
from google.colab import files
uploaded = files.upload()
dataset_zip = next(iter(uploaded.keys()))
dataset_zip
```

```python
import zipfile, os
os.makedirs('/content/removals_dataset', exist_ok=True)
with zipfile.ZipFile(dataset_zip, 'r') as z:
    z.extractall('/content/removals_dataset')
!find /content/removals_dataset -maxdepth 3 -name data.yaml -print
```

Set `DATA_YAML` to the printed `data.yaml` path.

```python
DATA_YAML = '/content/removals_dataset/data.yaml'  # change if needed
```

## 3. Train YOLO Model

```python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
results = model.train(data=DATA_YAML, epochs=80, imgsz=640, batch=16, name='removals_yolo')
```

## 4. Export

Export the best model. If TensorFlow.js export is available in your environment, use `format='tfjs'`. Otherwise export ONNX/SavedModel and convert in a second step.

```python
best_model = YOLO('/content/runs/detect/removals_yolo/weights/best.pt')
best_model.export(format='tfjs')
```
